```python
import pandas as pd
import numpy as np
import datetime
```

```python
a=pd.read_csv('BTC-USD.csv')
b=pd.read_csv('DCOILWTICO.csv')
c=pd.read_csv('DEXUSEU.csv')
d=pd.read_csv('GOLDAMGBD228NLBM.csv')
e=pd.read_csv('SP500.csv')
```

```python
b['DATE'] = pd.to_datetime(b.DATE)
```

```python
a.DATE = a.DATE.astype(str)
b.DATE = b.DATE.astype(str)
c.DATE = c.DATE.astype(str)
d.DATE = d.DATE.astype(str)
e.DATE = e.DATE.astype(str)
```

```python
bc=pd.merge(b,c, how='outer', on='DATE')
bcd=pd.merge(bc,d, how='outer', on='DATE')
bcde=pd.merge(bcd,e, how='outer', on='DATE')
bcde
```

```python
abcde=pd.merge(a,bcde, how='inner', on='DATE')
abcde.head()
```
